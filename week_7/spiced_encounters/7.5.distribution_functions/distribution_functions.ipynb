# Empirical Distribution

```python
def roll_dice(n=1, trials = 1):
    return np.random.randint(1,7,(trials,n))  # n : number of dices
```

```python
plt.hist(roll_dice(n=1, trials =100000).sum(axis=1), bins=60);
```

```python
plt.hist(roll_dice(n=10, trials =100000).sum(axis=1), bins=600);
```
