```python
import seaborn as sns
import matplotlib.pyplot as plt


cmap = sns.cubehelix_palette(rot=-.4, as_cmap=True, reverse =False)
df_subset = final_data.loc[final_data['year'] == 2000]
p =sns.scatterplot(data=df_subset, size = 'population', 
    alpha=.7, sizes = (30,350),hue="population",palette = 'seismic',
    x="life_expectancy", y="fertility_rate"
    )
p.legend(loc='center left', bbox_to_anchor=(0.9, 0.5), ncol=1)



```

```python
for year in range(1960,2015):
    df_subset = final_data.loc[final_data['year'] == year]
    p =sns.scatterplot(data=df_subset, size = 'population', 
        alpha=.6, sizes = (30,350),hue="population",palette = 'seismic',
        x="life_expectancy", y="fertility_rate"
        )
    p.legend(loc='center left', bbox_to_anchor=(0.9, 0.5), ncol=1)
    plt.ylim(0, 10)
    plt.xlim(25, 90)
    plt.axis((0,10,25,90))
    plt.savefig(f'lifeexp_{year}.png')
    plt.close()

```

```python
import imageio
import os

images = []

for i in range(1960, 2015):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=10)
for year in range(1960, 2015):
    os.remove(f'lifeexp_{year}.png')
```
