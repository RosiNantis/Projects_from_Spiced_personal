```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset = final_data.loc[final_data['year'] == 2000]
sns.scatterplot(x='life_expectancy', y='fertility_rate',
            data=df_subset, size = 'population', alpha=0.6)
```

```python
for year in range(1960,2015):
    df_subset = final_data.loc[final_data['year'] == year]
    sns.scatterplot(x='life_expectancy', y='fertility_rate',
        data=df_subset, size = 'population', alpha=0.6)
    plt.savefig(f'lifeexp_{year}.png')
    plt.close()

```

```python
import imageio
import os

images = []

for i in range(1960, 2015):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=40)
for year in range(1960, 2015):
    os.remove(f'lifeexp_{year}.png')
```
