```python
import pandas as pd
```

```python
df_expectancy= pd.read_excel('gapminder_lifeexpectancy.xlsx')
df_expectancy
```

```python
df_continents= pd.read_csv('continents.csv',sep = ';')

df_continents['country'].nunique()
df_continents
```

```python
final_df = pd.merge(left=df_expectancy, right=df_continents, left_on='Life expectancy',right_on='country', how='inner')
final_df
```

```python
# drop the two columns we used to create the id 
final_df.drop(['country'], axis=1, inplace=True)
final_df
```

```python
final = final_df.melt(
    id_vars = ["continent","Life expectancy"],
    var_name = "years",
    value_name = "expectancy")
final
```

```python
import seaborn as sns
_ = sns.boxplot(data=final, x='expectancy',y='continent')
#final[['expectancy','continent']].plot(kind='box')
```
