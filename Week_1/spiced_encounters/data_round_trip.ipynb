# Data Round-Trip

```python
import pandas as pd
```

## CSV

Read the CSV file `large_countries_2015.csv`

```python
read_file = pd.read_csv('./population.csv')
read_file
```

## Excel

* Write the data into an Excel file
* Read the data again

```python
read_file.to_excel('population.xlsx', sheet_name="Testing", index=False)
read_file_excel = pd.read_excel('./population.xlsx')
read_file_excel
```

## JSON - Java Script Object Notation

* Write the data into a JSON file
* Read the data again

```python
read_file_excel.to_json('population.json')
read_file_json = pd.read_json('population.json')
read_file_json
```

Create a connection to a new SQLite database:

```python
from sqlalchemy import create_engine
db = create_engine('sqlite:///mydatabase.db')
```

Write the data into the database:

```python
df = pd.DataFrame(read_file)
df.to_sql('population', db)
```

```python
import sqlite3

# Create a SQL connection to our SQLite database
con = sqlite3.connect("mydatabase.db")

cur = con.cursor()

# The result of a "cursor.execute" can be iterated over by row
for row in cur.execute('SELECT * FROM 1980;'):
    print(row)

# Be sure to close the connection
con.close()
```
